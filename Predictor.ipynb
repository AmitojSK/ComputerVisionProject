```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Activation, Dropout
from keras.optimizers import SGD
from keras.callbacks import ModelCheckpoint
```

```python
model = Sequential()
```

```python
Emodel = Sequential()
```

```python
Amodel = Sequential()
```

```python
model.add(Conv2D(200, (3,3), input_shape=(100, 100, 1)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(100, (3,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Dropout(0.3))
model.add(Flatten())

model.add(Dense(50, activation='relu'))
model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Emotions
Emodel.add(Conv2D(200, (3,3), input_shape=(100, 100, 1)))
Emodel.add(Activation('relu'))
Emodel.add(MaxPooling2D(pool_size=(2, 2)))

Emodel.add(Conv2D(100, (3,3)))
Emodel.add(Activation('relu'))
Emodel.add(MaxPooling2D(pool_size=(2, 2)))

Emodel.add(Dropout(0.3))
Emodel.add(Flatten())

Emodel.add(Dense(50, activation='relu'))
Emodel.add(Dense(3, activation='softmax'))

Emodel.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#Age
Amodel.add(Conv2D(200, (3,3), input_shape=(100, 100, 1)))
Amodel.add(Activation('relu'))
Amodel.add(MaxPooling2D(pool_size=(2, 2)))

Amodel.add(Conv2D(100, (3,3)))
Amodel.add(Activation('relu'))
Amodel.add(MaxPooling2D(pool_size=(2, 2)))

Amodel.add(Dropout(0.3))
Amodel.add(Flatten())

Amodel.add(Dense(50, activation='relu'))
Amodel.add(Dense(12, activation='softmax'))

Amodel.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
cp = ModelCheckpoint('model-best', verbose=0, save_best_only=True)
model.fit(X_train, Y_train, epochs = 5, callbacks=[cp], validation_split=0.2)
```

```python
#Age
cp = ModelCheckpoint('amodel-best', verbose=0, save_best_only=True)
Amodel.fit(AX_train, AY_train, epochs = 20, callbacks=[cp], validation_split=0.2)
```

```python
face_detect = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
```

```python
source = cv2.VideoCapture(0)
while 1:
    not_to_use, image = source.read()
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_detect.detectMultiScale(gray, 1.5, 1)
    for (x, y, w, h) in faces:
        face_roi = gray[y:y+w, x:x+h]
        resized_image = cv2.resize(face_roi, (100, 100))
        normalized_image = resized_image/255
        reshaped_face = np.reshape(normalized_image, (1, 100, 100, 1))
        result = mod.predict(reshaped_face)[0]
        eresult = emod.predict(reshaped_face)[0]
        aresult = amod.predict(reshaped_face)[0]
        k = aresult.argmin()

        if k==0:
            cv2.putText(image, "(0-2) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==1:
            cv2.putText(image, "(3-6) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==2:
            cv2.putText(image, "(7-11) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==3:
            cv2.putText(image, "(17-21) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==4:
            cv2.putText(image, "(12-16) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==5:
            cv2.putText(image, "(22-27) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==6:
            cv2.putText(image, "(28-35) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==7:
            cv2.putText(image, "(36-43) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==8:
            cv2.putText(image, "(44-50) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==9:
            cv2.putText(image, "(51-55) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==10:
            cv2.putText(image, "(56-60) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        elif k==11:
            cv2.putText(image, "(60+) years", (x, y+h+30), cv2.FONT_HERSHEY_SIMPLEX, 1, (230, 216, 173), 2)
        #print("Result : ", result)
        if result[0] >  result[1]:
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(image, "Male", (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

        if result[1] >  result[0]:
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)
            cv2.putText(image, "Female", (x, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

        if eresult[0] >  eresult[1]:
            if eresult[0] > eresult[2]:
                cv2.putText(image, "Neutral", (x+200, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (153, 255, 255), 2)
            else :
                cv2.putText(image, "Sad", (x+200, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (153, 255, 255), 2)


        else:
            if eresult[1] > eresult[2]:
                cv2.putText(image, "Happy", (x+200, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (153, 255, 255), 2)
            else :
                cv2.putText(image, "Sad", (x+200, y), cv2.FONT_HERSHEY_SIMPLEX, 1, (153, 255, 255), 2)


    cv2.imshow('Verzeo Project', image)
    key = cv2.waitKey(1)
    if key == 27:
        break
cv2.destroyAllWindows()
source.release()
```
